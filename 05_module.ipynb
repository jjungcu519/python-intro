### json변환 및 회차 변경

````python
payload = {
    'method' : 'getLottoNumber',
    'drwNo' : 1126,
}

r = requests.get('https://www.dhlottery.co.kr/common.do', params = payload)
#print(r.text)
#print(type(r.text)) #JSON과 DICT은 유사하지만 변환(convert)작업이 필요하다. -> str로 인식
#print(r.json())
#print(type(r.json()))

lotto_dict = r.json()
print(lotto_dict['drwtNo1'])
print(lotto_dict['drwtNo2'])
print(lotto_dict['drwtNo3'])
print(lotto_dict['drwtNo4'])
print(lotto_dict['drwtNo5'])
print(lotto_dict['drwtNo6'])
#print(r.text['drwtNo1'])
````

````python
#1 :이번주 당첨번호 불러오기
#2 : 로또 5번 구매하기 (랜덤 리스트 5개 뽑기)
#3 : 몇등인지 검증

payload = {
    'method' : 'getLottoNumber',
    'drwNo' : 1126,
}

r = requests.get('https://www.dhlottery.co.kr/common.do', params = payload)
lotto_dict = r.json()

this_week = [lotto_dict['drwtNo1'],lotto_dict['drwtNo2'],lotto_dict['drwtNo3'],lotto_dict['drwtNo4'],lotto_dict['drwtNo5'],lotto_dict['drwtNo6']]
bonus = [lotto_dict['bnusNo']]
print('이번주 당첨번호 :', this_week, 'BN:', bonus)


for i in range(10):
    numbers = list(range(1,46))
    r =random.sample(numbers, 6)

    result = 0
    result2 = 0
    
    for n in r:
        if n in this_week:
            result += 1
        elif n in bonus:
            result2 = 1
        
            
    print(f'{i+1}회차:', r)
    if result < 3:
        print('꽝')
    elif result == 3:
        print('5등')
    elif result == 4:
        print('4등')
    elif result == 5 and result2 == 0:
        print('3등')
    elif result == 5 and result2 == 1:
        print('2등')
    else:
        print('1등')
````
